```python
def print_to_file(filename, vector):
    ans = []
    for i in range(len(vector)):
        ans.append([i,vector[i]])
    with open(filename, "wb") as f:
        writer = csv.writer(f)
        writer.writerows(ans)
```

```python
def mean_matrix(v, n):
    v = np.matrix(v)
    mean = v.mean(axis=0)
    return np.repeat(mean,n,axis=0)
```

```python
def fit_spread(samples, cps, kernel='knn', alpha=1.0, n_neighbors = 2, gamma = 0.1):
    n, _ = combined_data.shape
    labels = -np.ones(n)
    labels[cps[:,0]] = cps[:,1]
    label_spread = label_propagation.LabelSpreading(kernel=kernel, alpha=alpha, gamma=gamma, n_neighbors=n_neighbors)
    label_spread.fit(samples, labels)
    labels = label_spread.transduction_
    
    plt.figure(num=None, figsize=(15, 8), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(samples[:, 0], samples[:, 1], c=labels, cmap=plt.cm.spectral)
    return labels
```
